```python
import os
from glob import glob
import shutil
import pandas as pd
import numpy
import toml
```

```python
data_dir = "/home/yiting/Documents/Data/Videos/Test/raw"
preprocessed_dir = "/home/yiting/Documents/Data/Videos/Test/preprocessed"
camera_views = ["To", "TL", "TR", "BL", "BR"]
subject_name = 'Neo'
```

```python
trials = sorted(os.listdir(data_dir))
for t in trials:
    trialname_parts = t.split('_')
    for c in camera_views:
        raw_vid_file = os.path.join(data_dir, t, "cam" + c +".mp4")
        new_vid_name = subject_name + "_" + trialname_parts[0] + "_" + trialname_parts[1] + "_" + "cam" + c + ".mp4"
        lp_vid_file = os.path.join(preprocessed_dir, new_vid_name)
        shutil.copyfile(raw_vid_file, lp_vid_file)
```
