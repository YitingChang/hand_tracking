```python
import os
import shutil
from pathlib import Path
```

```python
video_trial_name = "Neo_2025-12-21_11-59-32"
lp_project_dir = Path('/home/yiting/LPProjects/multiview_2025')
model_folder_dir = lp_project_dir / 'models'
camera_names = ['camTo', 'camTR', 'camTL', 'camBL', 'camBR']
eks_data_dir = Path('/home/yiting/Documents/GitHub/eks/data')
eks_video_dir = eks_data_dir / video_trial_name
os.makedirs(eks_video_dir, exist_ok=True)
```

```python
model_folder_names = sorted([m.name for m in model_folder_dir.iterdir() if m.is_dir()])

for model_name in model_folder_names:
    video_pred_dir = model_folder_dir / model_name / 'video_preds'
    for c in camera_names:
        camera_pred_dir = video_pred_dir / f'{video_trial_name}_{c}.csv'
        if camera_pred_dir.is_file():
            new_file_name = f'{video_trial_name}_{c}_{model_name}.csv'
            dest_path = eks_video_dir / new_file_name
            shutil.copy(camera_pred_dir, dest_path)
```
