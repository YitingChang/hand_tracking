```python
# Setting
data_dir = r"/media/yiting/NewVolume/Data/Videos"
output_dir = r"/media/yiting/NewVolume/Analysis/example_frames"
session_folder_paths = sorted(glob(os.path.join(data_dir,"2025-12*")))
camera = 'camTL'
for session_path in session_folder_paths:
    session_name = Path(session_path).name
    trial_folder_paths = sorted(glob(os.path.join(session_path, "cameras", "*")))
    os.makedirs(os.path.join(output_dir, session_name), exist_ok=True)
    for trial_folder_path in trial_folder_paths:
        trial_name = Path(trial_folder_path).name
        new_trial_name = "_".join(trial_name.split("_")[:2])
        video_path = os.path.join(trial_folder_path, f"{camera}.mp4")
        output_path = os.path.join(output_dir, session_name, f"{new_trial_name}_{camera}_frame300.png")

        cap = cv2.VideoCapture(video_path)
        # Frame index is 0-based â†’ 300th frame = index 299
        cap.set(cv2.CAP_PROP_POS_FRAMES, 299)

        ret, frame = cap.read()
        if not ret:
            raise RuntimeError("Failed to read frame 300")

        cv2.imwrite(output_path, frame)
        cap.release()

```
