```python
import os
import shutil
import cv2
```

```python
# Downsample images from images
# input: images
# output: images
```

```python
downsample = 300
session_folder = r'E:\Hand_tracking\Calibration\2023-12-13_18-15-30_620762'
downsample_folder = r'E:\Hand_tracking\Calibration\2023-12-13_downsampled_300'
cameras = os.listdir(session_folder)
for c in cameras:
    image_folder = os.path.join(session_folder, c) # folder containing images
    images = sorted(os.listdir(image_folder))
    os.makedirs(os.path.join(downsample_folder, c), exist_ok = True)
    for img in images:
        if int(img[7:13]) % downsample == 0:
            src = os.path.join(image_folder, img)
            dst = os.path.join(downsample_folder, c, img)
            shutil.copy(src, dst)
 
```

```python
# JARVIS intrinsic calibration
# session_folder = r'/home/yiting/Documents/Data/Videos/2024-03-31/2024-03-31/cameras/2024-03-31_11-09-47_788919'
session_folder = r'/home/yiting/Documents/Data/Videos/2024-03-31/2024-03-31/cameras/2024-03-31_11-25-40_283269'
intrinsic_folder = r"/home/yiting/Documents/Jarvis_projects/Calibration/Intrinsics"
camera_name = 'camBR'
video_start = 1 # in second
video_end =  65 # in second
downsample = 10

video_path = os.path.join(session_folder, camera_name + '.mp4')
vidcap = cv2.VideoCapture(video_path)
frame_rate = vidcap.get(cv2.CAP_PROP_FPS)
width  = int(vidcap.get(cv2.CAP_PROP_FRAME_WIDTH))   # float `width`
height = int(vidcap.get(cv2.CAP_PROP_FRAME_HEIGHT))  # float `height`
frame_start = int(video_start*frame_rate) 
frame_end = int(video_end*frame_rate)

video_save_path = os.path.join(intrinsic_folder, camera_name + '.mp4')
video = cv2.VideoWriter(video_save_path,  #Provide a file to write the video to
            fourcc = cv2.VideoWriter_fourcc(*'mp4v'), # code for mp4
            fps=int(frame_rate/downsample),                #How many frames do you want to display per second in your video?
            frameSize=(width, height))                #The size of the frames you are writing

for idx in range(frame_start, frame_end, downsample):
    vidcap.set(cv2.CAP_PROP_POS_FRAMES, idx) # Set the frame to get
    ref, frame = vidcap.read()
    video.write(frame)

cv2.destroyAllWindows()
video.release()
```

```python
# JARVIS extrinsic calibration
# session_folder = r'/home/yiting/Documents/Data/Videos/2024-03-31_calibration/2024-03-31/cameras/2024-03-31_11-09-47_788919'
# session_folder = r'/home/yiting/Documents/Data/Videos/2024-03-31_calibration/2024-03-31/cameras/2024-03-31_11-25-40_283269'
session_folder = r'/home/yiting/Documents/Data/Videos/2024-03-31_calibration/2024-03-31/cameras/2024-03-31_11-21-14_305219'
intrinsic_folder = r"/home/yiting/Documents/Jarvis_projects/Calibration/Extrinsics"
camera_pair = ['camTo', 'camTR']
video_start = 1 # in second
video_end =  65 # in second
downsample = 10

for camera_name in camera_pair:
    video_path = os.path.join(session_folder, camera_name + '.mp4')
    vidcap = cv2.VideoCapture(video_path)
    frame_rate = vidcap.get(cv2.CAP_PROP_FPS)
    width  = int(vidcap.get(cv2.CAP_PROP_FRAME_WIDTH))   # float `width`
    height = int(vidcap.get(cv2.CAP_PROP_FRAME_HEIGHT))  # float `height`
    frame_start = int(video_start*frame_rate) 
    frame_end = int(video_end*frame_rate)

    video_save_path = os.path.join(intrinsic_folder, camera_pair[0] + '-' + camera_pair[1], camera_name + '.mp4')
    video = cv2.VideoWriter(video_save_path,  #Provide a file to write the video to
                fourcc = cv2.VideoWriter_fourcc(*'mp4v'), # code for mp4
                fps=int(frame_rate/downsample),                #How many frames do you want to display per second in your video?
                frameSize=(width, height))                #The size of the frames you are writing

    for idx in range(frame_start, frame_end, downsample):
        vidcap.set(cv2.CAP_PROP_POS_FRAMES, idx) # Set the frame to get
        ref, frame = vidcap.read()
        video.write(frame)

    cv2.destroyAllWindows()
    video.release()
```

```python
# Anipose calibration: Video concatenation and downsample
downsample = 100
anipose_folder = r'/home/yiting/Documents/Anipose_projects/Anipose_6cam_240331/Anipose_240701/calibration_videos_downsampled_100'
session_folder = r'/home/yiting/Documents/Data/Videos/2024-03-31_calibration/2024-03-31/cameras'
trials = os.listdir(session_folder)
cameras = os.listdir(os.path.join(session_folder, trials[0]))
# Get frame rate and frame size
video_path = os.path.join(session_folder, trials[0], cameras[0])
vidcap = cv2.VideoCapture(video_path)
frame_rate = vidcap.get(cv2.CAP_PROP_FPS)
width  = int(vidcap.get(cv2.CAP_PROP_FRAME_WIDTH))   # float `width`
height = int(vidcap.get(cv2.CAP_PROP_FRAME_HEIGHT))  # float `height`

for c in cameras:   
    video_save_path = os.path.join(anipose_folder, c[:5] + '.mp4')
    video = cv2.VideoWriter(video_save_path,  #Provide a file to write the video to
                fourcc = cv2.VideoWriter_fourcc(*'mp4v'), # code for mp4
                fps=int(frame_rate/downsample),                #How many frames do you want to display per second in your video?
                frameSize=(width, height))                #The size of the frames you are writing
    for t in trials:
        video_path = os.path.join(session_folder, t, c)
        vidcap = cv2.VideoCapture(video_path)
        total_frame_number = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))

        for idx in range(total_frame_number):
            if idx % downsample == 0:
                vidcap.set(cv2.CAP_PROP_POS_FRAMES, idx) # Set the frame to get
                ref, frame = vidcap.read()
                video.write(frame)

    cv2.destroyAllWindows()
    video.release()
```
